```python
### Task 1 — Data Parsing & Profile Cleaning 
print("""-----Task 1 — Data Parsing & Profile Cleaning-----   
      """)
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

#Cleaning the data
for student in raw_students:
    student["name"] = student["name"].strip() #Removing leading and trailing white spaces in the name
    student["name"] = student["name"].title() #Converting the name to title case
    student["roll"] = int(student["roll"])    #Converting the roll number from string to integer
    student["marks_str"] = student["marks_str"].split(",") # Splitting the marks string into a marks list
    student["marks_str"] = [int(x.strip()) for x in student["marks_str"] ]     # Loops through the marks list, strips the extra spaces in the string and then converts it into integer

#Verifying if the name is valid
    name_nospace = student["name"].replace(" ","")  #Removing spaces in the name
    if(name_nospace.isalpha()):                     #Checking if the name contains only alphabetic characters
        print(student["name"],": ✓ Valid name")
    else:
        print(student["name"],": ✗ Invalid name")

#Printing a formatted profile card for each student using f-strings
    print(f"""================================
Student : {student["name"]}
Roll No : {student["roll"]}
Marks   : {student["marks_str"]}
================================
""");

#Printing the name in ALL CAPS and lowercase for a student whose roll number is 103
for student in raw_students:
    if(student["roll"] == 103):
        print('Name of the student with roll number 103')
        print("Name in ALL CAPS:",student["name"].upper())
        print("Name in lowercase:",student["name"].lower());



### Task 2 — Marks Analysis Using Loops & Conditionals

print("""
-----Task 2 - Marks Analysis Using Loops & Conditionals-----
      """)
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
total_marks = 0
average_marks = 0
highest_marks = marks[0]
highest_scoring_subject = ""
lowest_marks = marks[0]
lowest_scoring_subject = ""

#Printing Subject, Marks and Grade
for i in range(len(subjects)):
    if (marks[i]>= 90 and marks[i]<=100):                                    #Marks Range:90-100, Grade: A+
        print("Subject:",subjects[i],", Marks:", marks[i],", Grade:","A+")   
    elif (marks[i]>= 80 and marks[i]<=89):                                   #Marks Range:80-89, Grade: A
         print("Subject:",subjects[i],", Marks:", marks[i],", Grade:","A")  
    elif (marks[i]>= 70 and marks[i]<=79):                                   #Marks Range:70-79, Grade: B
         print("Subject:",subjects[i],", Marks:", marks[i],", Grade:","B")
    elif (marks[i]>= 60 and marks[i]<=69):                                   #Marks Range:60-69, Grade: C
         print("Subject:",subjects[i],", Marks:", marks[i],", Grade:","C")
    else:                                                                    #Marks Range: Below 60, Grade: F
         print("Subject:",subjects[i],", Marks:", marks[i],", Grade:","F")

    total_marks = total_marks + marks[i]

    if(marks[i]>highest_marks):
        highest_marks = marks[i]
        highest_scoring_subject = subjects[i]


    if(marks[i] < lowest_marks):
        lowest_marks = marks[i]
        lowest_scoring_subject = subjects[i]

print("\nTotal Marks = ",total_marks);    #Total Marks of all subjects

average_marks = total_marks / len(marks);               #Average Marks 
print(f"Average Marks = {average_marks:.2f}");          #Rounding to 2 decimal points

#Highest & Lowest Scoring Subject with marks
print(f"""Highest Scoring Subject is {highest_scoring_subject} , Highest Marks = {highest_marks} 
Lowest Scoring Subject is {lowest_scoring_subject} , Lowest Marks = {lowest_marks}""");

#Marks-entry system that allows adding new subjects
new_subject_name = input("Enter Subject Name: ")   
new_subject_marks=0
total_new_subjects = 0
new_total_marks = 0

while(new_subject_name != 'done'):
    new_marks = input("Enter marks for the subject: ")
    try:                                                                            
        new_subject_marks = int(new_marks)                                           #Handling Invalid Marks Inputs : non-numeric values
        if(0 <= new_subject_marks <= 100):                         #Handling Invalid Marks Inputs : numbers outside 0–100) 
            subjects.append(new_subject_name)                                               # Adding new subject
            marks.append(new_subject_marks)                                                 # Adding marks of the newly added subject
            total_new_subjects = total_new_subjects + 1
            new_total_marks = new_total_marks + new_subject_marks
            new_subject_name = input("Enter Subject Name: ")
        else:
            print("do not crash, and do not add invalid entries to the list.") 
    except ValueError:
        print("do not crash, and do not add invalid entries to the list.")
    
print(f"\nTotal New Subjects Added = {total_new_subjects}");                          #Printing number of newly added subjects
updated_total = total_marks + new_total_marks
#updated_subject_count = len(marks)
updated_average_marks = updated_total / len(marks)
print(f"Updated average marks across all subjects = {updated_average_marks:.2f}")   #Printing updated average across all subjects
    
    
### Task 3: Class Performance Summary

print("""

-----Task 3 - Class Performance Summary-----
      """)

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# Computing average of the students, Assigning result status as pass/fail
class_details = []
pass_count = 0
fail_count = 0
for student in class_data:
    student_marks = student[1]
    total_student_marks = 0
    for marks in student_marks:
        total_student_marks = total_student_marks + marks
    average_student_marks = round(total_student_marks / len(student_marks),2)              # Computing Average and rounded to 2 decimal places 
    if(average_student_marks >= 60):                                                       
        result_status = "Pass"                                                             # Assigning 'Pass' result for students whose average is >=60
        pass_count = pass_count + 1                                                        # Counting Number of Students who have passed
    else:
        result_status = "Fail"                                                             # Assigning 'Pass' result for students whose average is below 60
        fail_count = fail_count + 1                                                        # Counting Number of Students who have failed
    
    student_dict = {
        "name": student[0],
        "average": average_student_marks,
        "result": result_status
    }
    class_details.append(student_dict)

# Printing Formatted Class Report
print(f"{"Name":<20} | {"Average":<10} | {"Status":<10}")                                   
print("------------------------------------------");
for student in class_details:
    print(f"""{student["name"]:<20} | {student["average"]:<10.2f} | {student["result"]:<10}""");

#Printing No. of students who have passed/failed, Class Topper details and Class Average
print("\nNumber of students who passed : ",pass_count)
print("Number of students who failed : ",fail_count)

topper = class_details[0]
sum_class_average = 0
for student in class_details:
    if(student["average"] > topper["average"]):
        topper = student
    sum_class_average = sum_class_average + student["average"]
print(f"Class Topper: {topper["name"]}, Average : {topper["average"]:.2f}");  #Finding Class Topper Name and Average
print(f"Class Average: {sum_class_average/len(class_details):.2f}")        #Calculating Class Average

### Task 4: String Manipulation Utility

print("""

-----Task 4 - String Manipulation Utility -----
      """)

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

#1. Stripping leading and trailing white space
clean_essay = essay.strip();
print(f"""#1. Stripping leading and trailing white spaces:
{clean_essay}""");

#2. Converting to title case
print(f"""\n#2. Title case: 
{clean_essay.title()}""");

#3. Number of times the word 'python' appears
print(f"\n#3. Number of times the word 'python' appears : {clean_essay.count("python")}");

#4. Replacing 'python' with 'Python'
print(f"""\n#4. Replaced the word 'python' with 'Python':
{clean_essay.replace("python","Python")}""");

#5. Splitting into Individual Sentences
print(f"""\n#5. Individual Sentences:
{clean_essay.split(". ")}""");

#6. Printing each sentence on a new line, numbered and ending with a dot(.)
print("\n#6. Each sentence on a new line, numbered with a period(.):")
num = 1
for sentence in clean_essay.split(". "):
    if "." not in sentence:
       sentence = sentence + "."
                             
    print(f"{num}. {sentence}")
    num = num + 1

```
